### 1. K-NN model
##### Use K-NN to model

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

#### 1.1 K-NN using selected dataset
##### as the original dataset contain too much dimension after embedding, use the selected dataset.

```python
# first use the basic no text data
data = pd.read_csv("../COMP30027_2024_asst2_data/train_dataset_select.csv")
# drop id before calculation, add back after
data = data.drop(columns='id')
```

```python
# normalize the data
X = data.drop(columns=["imdb_score_binned"])
y = data["imdb_score_binned"]

# set up scalers
mmscaler = MinMaxScaler()
ssscaler = StandardScaler()

mmscaler.fit(X)
ssscaler.fit(X)

mmX = mmscaler.transform(X)
ssX = ssscaler.transform(X)
```

```python
# cross validation with different k
k_values = range(1, 51, 5)
mm_accuracies = []
ss_accuracies = []

# Loop over each k value
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)

    # cross validation for minmax scaled data
    scores = cross_val_score(knn, mmX, y, cv=10, scoring='accuracy')
    mm_accuracies.append(scores.mean())

    # cross validation for standard scaled data
    scores = cross_val_score(knn, ssX, y, cv=10, scoring='accuracy')
    ss_accuracies.append(scores.mean())
```

```python
print("Maximum accuracy for min-max normalization", max(mm_accuracies))
print("Maximum accuracy for standardization normalization", max(ss_accuracies))
```

```python
# plot the output
plt.figure(figsize=(10, 6))
plt.plot(k_values, mm_accuracies, marker='o', linestyle='-', color='b', label='MinMax Scaling')
plt.plot(k_values, ss_accuracies, marker='s', linestyle='--', color='r', label='Standard Scaling')
plt.xlabel('Number of Neighbors: k')
plt.ylabel('Accuracy')
plt.title('KNN Accuracy with Different k Values and Scaling Methods')
plt.legend()
plt.grid(True)
plt.savefig("../COMP30027_2024_asst2_graph/KNN_Accuracy_No_Text.png")
plt.show()
```

```python
# lets use the best performing k to analysis
knn = KNeighborsClassifier(n_neighbors=15)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=715)
knn.fit(X_train, y_train)
predictions = knn.predict(X_test)
```

```python
cm = confusion_matrix(y_test, predictions)
# plot the confusion matrix
plt.figure(figsize=(8, 6))

sns.heatmap(cm, annot=True, fmt="d", cmap="Purples", cbar=False)

plt.title('K-NN Confusion Matrix K=15')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.savefig("../COMP30027_2024_asst2_graph/KNN_Confustion_Matrix.png")
plt.show()
```

This model does not achieve very well accuracy, need to try other models.\
One reason of this model is not suitable for this dataset might be the dataset contains quite a lot categorical data.\
And K-NN's distance calculation way of prediction might not capture the underlying relation of the data with the label.\
