### 5. Neural networks
##### Explore varies neural networks using keras library.

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MultiLabelBinarizer, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay

import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Input, Embedding, Flatten, Dense, concatenate
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.text import Tokenizer

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tf.random.set_seed(715)
```

```python
# first use the basic no text data
data = pd.read_csv("../COMP30027_2024_asst2_data/train_dataset_select.csv")
test_data = pd.read_csv("../COMP30027_2024_asst2_data/test_dataset_select.csv")
# drop id before calculation, add back after
data = data.drop(columns='id')
test_data = test_data.drop(columns='id')
```

```python
# split the data
X = data.drop(columns=["imdb_score_binned"])
y = data["imdb_score_binned"]



X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=715)

# set up scalers
mmscaler = MinMaxScaler()
ssscaler = StandardScaler()


mmX_train = mmscaler.fit_transform(X_train)
mmX_test = mmscaler.transform(X_test)
ssX_train = ssscaler.fit_transform(X_train)
ssX_test = ssscaler.transform(X_test)

```

```python
# design nn
model = Sequential([
    Dense(32, input_shape=(X_train.shape[1],), activation='relu'),
    Dense(16, activation='sigmoid'),
    Dense(y.nunique(), activation='softmax')
])

# instantiate
model.compile(optimizer='nadam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(ssX_train, y_train, epochs=50, batch_size=32, validation_split=0.2, callbacks=[early_stopping], shuffle=True)


# get accuracy
test_loss, test_acc = model.evaluate(ssX_test, y_test)

predictions = model.predict(ssX_test)
predictions = np.argmax(predictions, axis=1)
```

```python
# accuracy
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)

# precision and recall
precision = precision_score(y_test, predictions, average='weighted')
recall = recall_score(y_test, predictions, average='weighted')

print("Precision:", precision)
print("Recall:", recall)

# display the confusion matrix
cm = confusion_matrix(y_test, predictions)
print("Confusion Matrix:\n", cm)
```

```python
# plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="coolwarm", cbar=False, vmin=0, vmax=cm.max())
plt.title('Dense NN Confusion Matrix with feature selection')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.savefig("../COMP30027_2024_asst2_graph/NN_Confustion_Matrix_Select.png")
plt.show()
```

#### Kaggle prediction on test data

```python
sstest_data = ssscaler.transform(test_data)
```

```python
test_pred = model.predict(sstest_data)

ids = range(1, len(test_pred) + 1)
prediction = pd.DataFrame({'id': ids, 'imdb_score_binned': np.argmax(test_pred, axis=1)})
print(prediction['imdb_score_binned'].value_counts())
prediction.to_csv("../Kaggle_submissions/dense_nn_select.csv", index=False)
```

#### 5.2 using embedding layer to leverage the performance by adding removed feature that might contain information

Through this basic dense neural network, model achieve decent outcome, 
however, it does not utilize the powerful model of neural network's capibility of handling more complex model.
Hence, try add embedding layer to add the categorical data that we removed to add more feature into the model fitting.

```python
# same X, Y, train test
# grab the genre column
origin_data = pd.read_csv("../COMP30027_2024_asst2_data/train_dataset.csv")

# split genres and encode
origin_data['genres'] = origin_data['genres'].apply(lambda x: x.split('|'))
mlb = MultiLabelBinarizer()
genres_encoded = mlb.fit_transform(origin_data['genres'])

genres_train, genres_test = train_test_split(genres_encoded, test_size=0.2, random_state=715)
```

```python
# setup model

numerical_input = Input(shape=(ssX_train.shape[1],), name='num_input')
genre_input = Input(shape=(genres_encoded.shape[1],), name='genre_input')

concat = concatenate([numerical_input, genre_input])

dense = Dense(64, activation='relu')(concat)
output = Dense(y.nunique(), activation='softmax')(dense)

model = Model(inputs=[numerical_input, genre_input], outputs=output)
model.compile(optimizer='nadam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
history = model.fit([ssX_train, genres_train], y_train, epochs=50, batch_size=32, validation_split=0.2, callbacks=[early_stopping], shuffle=True)

test_loss, test_acc = model.evaluate([ssX_test, genres_test], y_test)

predictions = model.predict([ssX_test, genres_test])
predictions = np.argmax(predictions, axis=1)
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.savefig("../COMP30027_2024_asst2_graph/NN_Learning_Curve_with_genre.png")
plt.show()
```

```python
# accuracy
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)

# precision and recall
precision = precision_score(y_test, predictions, average='weighted')
recall = recall_score(y_test, predictions, average='weighted')

print("Precision:", precision)
print("Recall:", recall)

# display the confusion matrix
cm = confusion_matrix(y_test, predictions)
print("Confusion Matrix:\n", cm)
```

```python
# plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="coolwarm", cbar=False, vmin=0, vmax=cm.max())
plt.title('Dense NN Confusion Matrix with Genres Added')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.savefig("../COMP30027_2024_asst2_graph/NN_Confustion_Matrix_with_genre.png")
plt.show()
```
