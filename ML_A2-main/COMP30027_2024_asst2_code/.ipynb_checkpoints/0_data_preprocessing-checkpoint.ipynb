## 0. Data preprocess

##### Since dataset contains many different types of features, require preprocess the data.

```python
# import useful library
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import mutual_info_regression
from sklearn.feature_selection import f_classif
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MultiLabelBinarizer

import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import kstest, norm
```

```python
# import datasets
train_data = pd.read_csv("../COMP30027_2024_asst2_data/train_dataset.csv")
test_data = pd.read_csv("../COMP30027_2024_asst2_data/test_dataset.csv")


# count CountVectorizer features for director names, actor names, and possibly other text features
cv_path = "../COMP30027_2024_asst2_data/features_countvec/"
cvtrain_director_name = np.load(cv_path + "train_countvec_features_director_name.npy")
cvtrain_actor_1_name = np.load(cv_path + "train_countvec_features_actor_1_name.npy")
cvtrain_actor_2_name = np.load(cv_path + "train_countvec_features_actor_2_name.npy")

cvtest_director_name = np.load(cv_path + "test_countvec_features_director_name.npy")
cvtest_actor_1_name = np.load(cv_path + "test_countvec_features_actor_1_name.npy")
cvtest_actor_2_name = np.load(cv_path + "test_countvec_features_actor_2_name.npy")

# Doc2Vec features for plot keywords, genres, and other text features
d2v_path = "../COMP30027_2024_asst2_data/features_doc2vec/"
d2vtrain_plot = np.load(d2v_path + "train_doc2vec_features_plot_keywords.npy")
d2vtrain_genre = np.load(d2v_path + "train_doc2vec_features_genre.npy")
d2vtest_plot = np.load(d2v_path + "test_doc2vec_features_plot_keywords.npy")
d2vtest_genre = np.load(d2v_path + "test_doc2vec_features_genre.npy")

# FastText embeddings for movie titles
ft_path = "../COMP30027_2024_asst2_data/features_fasttext/"
fttrain_title = np.load(ft_path + "train_fasttext_title_embeddings.npy")
fttest_title = np.load(ft_path + "test_fasttext_title_embeddings.npy")
```

```python
# convert the numpy array to dataframes
def convert_to_dataframe(np_array, prefix, index):
    # convert nparray to df
    col_names = [f"{prefix}_feat_{i}" for i in range(np_array.shape[1])]
    return pd.DataFrame(np_array, columns=col_names, index=index)

df_cvtrain_director_name = convert_to_dataframe(cvtrain_director_name, 'director', train_data.index)
df_cvtrain_actor_1_name = convert_to_dataframe(cvtrain_actor_1_name, 'actor_1', train_data.index)
df_cvtrain_actor_2_name = convert_to_dataframe(cvtrain_actor_2_name, 'actor_2', train_data.index)

df_d2vtrain_plot = convert_to_dataframe(d2vtrain_plot, 'plot_keywords', train_data.index)
df_d2vtrain_genre = convert_to_dataframe(d2vtrain_genre, 'genres', train_data.index)

df_fttrain_title = convert_to_dataframe(fttrain_title, 'title', train_data.index)

df_cvtest_director_name = convert_to_dataframe(cvtest_director_name, 'director', test_data.index)
df_cvtest_actor_1_name = convert_to_dataframe(cvtest_actor_1_name, 'actor_1', test_data.index)
df_cvtest_actor_2_name = convert_to_dataframe(cvtest_actor_2_name, 'actor_2', test_data.index)

df_d2vtest_plot = convert_to_dataframe(d2vtest_plot, 'plot_keywords', test_data.index)
df_d2vtest_genre = convert_to_dataframe(d2vtest_genre, 'genres', test_data.index)

df_fttest_title = convert_to_dataframe(fttest_title, 'title', test_data.index)
```

As the dataset contains missing values, fill it with a reasonable guess.

```python
# handle missing values
missing_counts = train_data.isnull().sum()
print(missing_counts)
missing_row = train_data[train_data['language'].isnull()]
# as the language of a movie called "Labor Day" that the country is USA, good chance English
train_data.loc[2787, 'language'] = 'English'
```

From further inspect, the dataset also contains quite a bit 0 in the facebook like.\
Those are likely be missing values, and can alter the result quite a bit.\
Test the distribution of these column to decide what value should fill.

```python
# director_facebook_likes
print((train_data['director_facebook_likes'] == 0).sum())

modified_data = train_data['director_facebook_likes'].replace(0, np.nan)
filtered_data = modified_data.dropna()

stat, p = kstest(filtered_data, 'norm', args=(filtered_data.mean(), filtered_data.std()))
print('k-stat:', stat, 'p-value:', p)

# non normal distribution, use median to fill all 0.
median = train_data['director_facebook_likes'][train_data['director_facebook_likes'] != 0].median()
train_data['director_facebook_likes'] = train_data['director_facebook_likes'].replace(0, median)

median = test_data['director_facebook_likes'][test_data['director_facebook_likes'] != 0].median()
test_data['director_facebook_likes'] = test_data['director_facebook_likes'].replace(0, median)
```

```python
# movie_facebook_likes
print((train_data['movie_facebook_likes'] == 0).sum())

modified_data = train_data['movie_facebook_likes'].replace(0, np.nan)
filtered_data = modified_data.dropna()

stat, p = kstest(filtered_data, 'norm', args=(filtered_data.mean(), filtered_data.std()))
print('k-stat:', stat, 'p-value:', p)

# non normal distribution, use median to fill all 0.
median_value = train_data['movie_facebook_likes'][train_data['movie_facebook_likes'] != 0].median()
train_data['movie_facebook_likes'] = train_data['movie_facebook_likes'].replace(0, median_value)

median_value = test_data['movie_facebook_likes'][test_data['movie_facebook_likes'] != 0].median()
test_data['movie_facebook_likes'] = test_data['movie_facebook_likes'].replace(0, median_value)
```

The other numerical dataset contains minimum to no 0's, likely not missing.

Inspect label distribution:

```python
value_counts = train_data['imdb_score_binned'].value_counts()

# check precentage
total = value_counts.sum()
percentages = (value_counts / total) * 100

# sort with frequency
value_counts = value_counts.sort_values(ascending=False)
categories = value_counts.index.tolist()
value_counts.index = pd.Categorical(value_counts.index, categories=categories, ordered=True)

# plot
plt.figure(figsize=(10, 6))
sns.barplot(x=value_counts.index, y=value_counts.values, hue=value_counts.index, palette='coolwarm', dodge=False, legend=False)

# annotation
for i, (count, percentage) in enumerate(zip(value_counts.values, percentages)):
    plt.text(i, count + total * 0.01, f'{percentage:.2f}%', ha='center')

plt.title('Distribution of IMDB Score Binned')
plt.xlabel('IMDB Score Binned')
plt.ylabel('Count')
plt.savefig("../COMP30027_2024_asst2_graph/Score_Distribution.png")
plt.show()
```

Very skewed dataset.

#### 0.2 dataset with feature engineering and feature selection.
Since the full dataset contains a lot of dimensions, need to examine the features.

```python
train_f = train_data.copy()
test_f = test_data.copy()
```

#### Let's deal with columns one by one.
##### First we have **director name**, let's examine the feature first.

```python
director_counts = train_f['director_name'].value_counts()
print(director_counts)
print(len(train_f['director_name'].unique()))
```

##### Now check each director's scoring on movie score.

```python
# find the average score for the top 10 director that filmed most movies.
top_directors = train_f['director_name'].value_counts().head(10).index

filtered_data = train_f[train_f['director_name'].isin(top_directors)]

average_scores = filtered_data.groupby('director_name')['imdb_score_binned'].mean().reset_index()

# plot it out
plt.figure(figsize=(12, 6))
sns.barplot(x='director_name', y='imdb_score_binned', data=average_scores)
plt.xticks(rotation=45)
plt.title('Average Score by Director')
plt.xlabel('Director Name')
plt.ylabel('Average Movie Score')
plt.savefig("../COMP30027_2024_asst2_graph/Average_Score_Director.png")
plt.show()
```

There are no significant difference in between them, and the rest directors are less significant as not much movie filmed.\
It is best to remove this feature.

```python
train_f = train_f.drop(columns=['director_name'])
```

##### Now work on the **actor names**.
With actor names, let's examine as a whole.

```python
all_actors = pd.concat([train_f['actor_1_name'], train_f['actor_2_name'], train_f['actor_3_name']])

print(len(all_actors.unique()))
```

Despite there are 3 actor per movie, 3859 different actors in 3003 movie is too diverse for it to be conclude.\
Remove these features.

```python
train_f = train_f.drop(columns=['actor_1_name', 'actor_2_name', 'actor_3_name'])
```

##### Now look at **genres**.
As there aren't much genre, examine each of them by looking at their scoring.

```python
# indivisual genre
exploded_genres = train_f.assign(genres=train_f['genres'].str.split('|')).explode('genres')

# check frequency
print(exploded_genres['genres'].value_counts())
# plot the average score per genre
average_scores = exploded_genres.groupby('genres')['imdb_score_binned'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='genres', y='imdb_score_binned', data=average_scores)
plt.xticks(rotation=45)
plt.title('Average IMDb Score Binned by Genre')
plt.xlabel('Genre')
plt.ylabel('Average IMDb Score Binned')
plt.savefig("../COMP30027_2024_asst2_graph/Average_Score_Genre.png")
plt.show()
```

There are no clear relation in the genre, as the average is quite the same.\
Remove.

```python
train_f = train_f.drop(columns=['genres'])
```

##### Now for the **plot_keyword**.

```python
# indivisual plot_keywords
exploded_plot_keywords = train_f.assign(plot_keywords=train_f['plot_keywords'].str.split('|')).explode('plot_keywords')
# check frequency
print(exploded_plot_keywords['plot_keywords'].value_counts())
```

##### Now for country.

```python
print(train_f['country'].value_counts())
```

Only a few country has produced significant amount of movies, let's test the average score on those.

```python
# calculated score by the top frequent country
top_country = train_f['country'].value_counts().head(9).index

filtered_data = train_f[train_f['country'].isin(top_country)]

average_scores = filtered_data.groupby('country')['imdb_score_binned'].mean().reset_index()

# plot it out
plt.figure(figsize=(12, 6))
sns.barplot(x='country', y='imdb_score_binned', data=average_scores)
plt.xticks(rotation=45)
plt.title('Average Score by Country')
plt.xlabel('Country')
plt.ylabel('Average Movie Score')
plt.savefig("../COMP30027_2024_asst2_graph/Average_Score_Country.png")
plt.show()
```

Now for **content_rating**.

```python
print(train_f['content_rating'].value_counts())
```

Too much categories, we can combine some.

```python
rating_map = {
    'R': 'Adult',
    'NC-17': 'Adult',
    'X': 'Adult',
    'PG-13': 'Parental Guidance',
    'PG': 'Parental Guidance',
    'G': 'General',
    'Not Rated': 'Not Rated',
    'Unrated': 'Not Rated',
    'Approved': 'Other',
    'M': 'Other',
    'Passed': 'Other',
    'GP': 'Other'
}

# map into 5 classes
train_f['content_rating'] = train_f['content_rating'].map(rating_map)
```

```python
print(train_f['content_rating'].value_counts())
```

```python
# calculated score by the top frequent content_rating

average_scores = train_f.groupby('content_rating')['imdb_score_binned'].mean().reset_index()

# plot it out
plt.figure(figsize=(12, 6))
sns.barplot(x='content_rating', y='imdb_score_binned', data=average_scores)
plt.xticks(rotation=45)
plt.title('Average Score by Content Rating')
plt.xlabel('Content Rating')
plt.ylabel('Average Movie Score')
plt.savefig("../COMP30027_2024_asst2_graph/Average_Score_Content.png")
plt.show()
```

The majority 2 has equivalent score, while no significant change on others.\
Remove.

```python
train_f = train_f.drop(columns=['content_rating'])
```
