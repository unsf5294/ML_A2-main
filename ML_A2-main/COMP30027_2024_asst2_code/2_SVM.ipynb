### 2. Support Vector Machine

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.svm import SVC
from sklearn.datasets import make_blobs
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
```

#### 2.1 SVM using the selected dataset
##### as SVM also involves distance measurement, using original data would have too much dimensionality
##### use selected dataset here

```python
# first use the basic no text data
data = pd.read_csv("../COMP30027_2024_asst2_data/train_dataset_select.csv")
# drop id before calculation, add back after
data = data.drop(columns='id')
```

```python
# split the data
X = data.drop(columns=["imdb_score_binned"])
y = data["imdb_score_binned"]

# set up scalers
mmscaler = MinMaxScaler()
ssscaler = StandardScaler()

mmscaler.fit(X)
ssscaler.fit(X)

mmX = mmscaler.transform(X)
ssX = ssscaler.transform(X)

svm = SVC()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=715)
ssscaler.fit_transform(X_train)
ssscaler.transform(X_test)

svm.fit(X_train, y_train)
predictions = svm.predict(X_test)
```

```python
cm = confusion_matrix(y_test, predictions)
# Calculate accuracy
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)

# precision and recall
precision = precision_score(y_test, predictions, average='weighted')
recall = recall_score(y_test, predictions, average='weighted')  
print("Precision:", precision)
print("Recall:", recall)

# plot the confusion matrix
plt.figure(figsize=(8, 6))

sns.heatmap(cm, annot=True, fmt="d", cmap="Purples", cbar=False)

plt.title('SVM Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.savefig("../COMP30027_2024_asst2_graph/SVM_Confustion_Matrix.png")
plt.show()
```
